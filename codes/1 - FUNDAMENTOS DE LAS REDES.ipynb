# Operadores

```python
G = nx.Graph([(0, 1), (0, 2), (1, 2)])
H = nx.Graph([(3, 4), (3, 5), (4, 5), (4, 6)])
U = nx.union(G, H, rename=("G", "H"))
```

```python
nx.draw(U, with_labels=True, font_weight='bold')
```

```python
U.add_edge('G1', 'H6')
```

```python
nx.draw(U, with_labels=True, font_weight='bold')
```

```python
G = nx.Graph([(0, 1), (0, 2)])
H = nx.Graph([(0, 1), (1, 2)])
R = nx.compose(G, H)
```

```python
nx.draw(R, with_labels=True, font_weight='bold')
```

# Distribuci√≥n de grado

```python
G = nx.erdos_renyi_graph(10000, 0.1)

degrees = [G.degree(n) for n in G.nodes()]

degree_counts = {}
for degree in degrees:
    if degree in degree_counts:
        degree_counts[degree] += 1
    else:
        degree_counts[degree] = 1

sorted_degrees = sorted(degree_counts.items())

x, y = zip(*sorted_degrees)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'bo-', markersize=8)
plt.title('Distribuci√≥n de Grado')
plt.xlabel('Grado')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```
